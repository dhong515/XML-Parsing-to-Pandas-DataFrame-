```python
import xmltodict
import pandas as pd
import datetime

def structure_data():
    xml_file_path = 'data-studies.xml'
    with open(xml_file_path, 'r') as f:
        data = f.read()

    data = xmltodict.parse(data)

    doi_list = []
    publicationdate_list = []
    title_list = []
    abstract_list = []

    for element in data['bibdataset']['item']:
        doi_list.append(element['bibrecord']['item-info']['itemidlist']['ce:doi'])

        publicationdate = element['bibrecord']['head']['source']['publicationdate']
        date_str = f"{publicationdate['year']}-{publicationdate['month']}-{publicationdate['day']}"
        date_obj = datetime.datetime.strptime(date_str, "%Y-%m-%d")
        date = date_obj.strftime("%Y-%m-%d")
        publicationdate_list.append(date)
        
        title_list.append(element['bibrecord']['head']['citation-title']['titletext']['#text'])
        
        abstract_list.append(element['bibrecord']['head']['abstracts']['abstract']['ce:para'])
    
    output = pd.DataFrame({
        'DOI': doi_list,
        'Publication Date': publicationdate_list,
        'Title': title_list,
        'Abstract': abstract_list
    })

    return output
    
def main():
    output = structure_data()
    display(output)

if __name__ == "__main__":
    main()
```
